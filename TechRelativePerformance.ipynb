# Tech Relative Performance

This notebook showcases the FANG stocks and other popular tech stocks relative performance to SPY.

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

**Config**

```python
sns.set_theme(palette="pastel", style=)
plt.rcParams["figure.figsize"] = (20,5)
```

**Stock tickers**

```python
tickers_to_plot = [
    "AAPL",
    "MSFT",
    "AMZN",
    "GOOG",
    "NVDA",
    "FB",
    "NFLX",
]
```

Get a ticker's price data.

```python
def get_ticker(ticker: str):
    ticker = yf.Ticker(ticker)
    data = ticker.history(period="1y", interval="1d", auto_adjust=False, actions=False)
    return data.dropna()
```

Plots a relative performance chart with the given tickers.

```python
def plot_relative_chart(base_ticker, comparitive_ticker, label=""):
    
    data = base_ticker["Close"] / comparitive_ticker["Close"]
    
    xpos = plt.np.arange(len(data))

    plt.plot(xpos, data, color='green')
    
    font = {'family':'serif','color':'purple','size':25}
    
    plt.title(label, fontdict=font)

    plt.show()
```

Getting price data of SPY

```python
spy = get_ticker("SPY")
spy
```

### Relative Performance Charts
Loops through all the provided tickers and plots them.

```python
for ticker in tickers_to_plot:
    
    data = get_ticker(ticker)
    plot_relative_chart(data, spy, ticker)
```
